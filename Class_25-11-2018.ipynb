```python
import numpy as np
persons = pd.DataFrame({
    'name': ['Shahzad', 'Taha', 'Fatima', 'Khani', 'Samia'],
    'age': np.random.randint(1,40, size=[5]),#[20, 22, 21,21, 3]
    'percent': np.random.randint(40,95, size=[5])
})
```

```python
persons
```

```python
def adult(age):
    if age < 18:
        return 'Child'
    return 'Adult'
```

```python
persons['status'] = persons['age'].apply(adult)
```

```python
persons
```

```python
def grade(per):
    if(per < 60):
        return 'F'
    elif(per < 70):
        return 'C'
    elif(per < 80):
        return 'B'
    elif(per < 90):
        return 'A'
    return 'A+'
```

```python
persons['grade'] = persons['percent'].apply(grade)
persons
```
