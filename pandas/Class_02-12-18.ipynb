Fuzzywuzzy Example

```python
def fuzEx(x):
    max = 0
    match = ''
    for i in data['Party']:
        r = f.ratio(x,i)
        if r > max:
            max = r
            match = i
    return match

corrected_data = data.Party.apply(fuzEx)
```
